```python
def trabalha_zip(arquivo):
    # Crie um objeto ZipFile e carrega arquivo nele
    with ZipFile(arquivo, 'r') as zipObj:
        # Obtenha uma lista de todos os nomes de arquivos arquivados do zip
        listOfFileNames = zipObj.namelist()
        # Iterar sobre os nomes de arquivo
        for fileName in listOfFileNames:
            # Verifica o nome do arquivo endswith o termo de interesse
            if fileName.endswith('Despesas_Empenho.csv'):
                # Extrai um único arquivo do zip para a pasta certa
                zipObj.extract(fileName, 'Despesas_Empenho')
                
            elif fileName.endswith('Despesas_Pagamento.csv'):
                # Extrai um único arquivo do zip para a pasta certa
                zipObj.extract(fileName, 'Despesas_Pagamento')
                
            elif fileName.endswith('Despesas_Pagamento_EmpenhosImpactados.csv'):
                # Extrai um único arquivo do zip para a pasta certa
                zipObj.extract(fileName, 'Despesas_Pagamento_EmpenhosImpactados')

    return 
```

```python
prefixo = "http://transparencia.gov.br/download-de-dados/despesas/2020"
```

```python
for vez in range(1, 9):
    mes = "0" + str(vez)
    print("Mês: ", mes)
    if mes == "01" or mes == "03" or mes == "07" or mes == "08" or mes == "05":
        for num in range(1, 32):
            if num < 10:
                dia = "0" + str(num)
            else:
                dia = str(num)
            link = prefixo + mes + dia
            print(link)
            
            try:
                wget.download(link)
            
                arquivo = "2020" + mes + dia + "_Despesas.zip"
            
                trabalha_zip(arquivo)
            except:
                print("ERRRO: ", link)
                pass
    
    elif mes == "04" or mes == "06" :
        for num in range(1, 31):
            if num < 10:
                dia = "0" + str(num)
            else:
                dia = str(num)
            
            link = prefixo + mes + dia
            print(link)
            
            try:
                wget.download(link)
            
                arquivo = "2020" + mes + dia + "_Despesas.zip"
            
                trabalha_zip(arquivo)
            except:
                print("ERRRO: ", link)
                pass
            
    elif mes == "02":
        for num in range(1, 30):
            if num < 10:
                dia = "0" + str(num)
            else:
                dia = str(num)
            
            link = prefixo + mes + dia
            print(link)
            
            try:
                wget.download(link)
            
                arquivo = "2020" + mes + dia + "_Despesas.zip"
            
                trabalha_zip(arquivo)
            except:
                print("ERRRO: ", link)
                pass
            
#    elif mes == "05":
#        for num in range(1, 12):
#            if num < 10:
#                dia = "0" + str(num)
#            else:
#                dia = str(num)
#            
#            link = prefixo + mes + dia
#            print(link)
#            
#            try:
#                wget.download(link)
#            
#                arquivo = "2020" + mes + dia + "_Despesas.zip"
#            
#                trabalha_zip(arquivo)
#            except:
#                print("ERRRO: ", link)
#                pass
```
